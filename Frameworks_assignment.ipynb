# **Part 2: Data Cleaning and Preparation (2-3 hours)**

````python
#checking the missing values
df.isnull().sum()
````

````python
#i will be removing the colums that are empty
df1 = df.drop(columns=['s2_id','arxiv_id','who_covidence_id','mag_id',])
````

````python
#Drops any row having a missing a value
df2 = df1.dropna(how='any')
````

````python
#Confirming whether there's missing value
df2.info()
````

````python
df2.head()
````

````python
#extracted a year in publish time then added a publication year in the dataset
df2 = df2.copy()
df2['Publication_Year'] = df2['publish_time'].dt.year
df2.head()
````

````python
#converting publication year datatype to int
df2['Publication_Year'] = df2['Publication_Year'].astype('Int64')
````

````python
#droping the publish time column in the dataset
df2 = df2.drop(columns=['publish_time'])
````

# **Part 3: Data Analysis and Visualization **

````python
#Counting  papers by publication year
papers_by_year = df2['Publication_Year'].value_counts().sort_index()
print(papers_by_year)
````

````python
#Identifying top journals publishing COVID-19 research
Top_journals = df2['journal'].value_counts().head(10)

print(Top_journals)
````

````python
#Finding most frequent words in titles (using simple word frequency)

from collections import Counter
import re

# I have a dataframe called df2 and I’m working with the 'title' column
titles = df2['title'].dropna().astype(str)

# I’m joining all the titles into one long string so I can process them together
all_text = " ".join(titles)

# I want all words in lowercase and only actual words (no punctuation or numbers)
words = re.findall(r'\b[a-zA-Z]+\b', all_text.lower())

# Here I’m counting how often each word appears
word_freq = Counter(words)

# I only care about the top 5 most frequent words
most_common_words = word_freq.most_common(5)

# Finally, I’m putting the results into a DataFrame so it’s easier to read
freq_df = pd.DataFrame(most_common_words, columns=['Word', 'Frequency'])
print(freq_df)
````

# ***visualizations***

````python
#Plot number of publications over time
plt.bar(papers_by_year.index, papers_by_year.values)
plt.xlabel("Publication Year")
plt.ylabel("Number of Publications")
plt.title("Number of Publications Over Time")
plt.show()
````

````python
#Create a bar chart of top publishing journals
plt.figure(figsize=(24, 11))
plt.bar(Top_journals.index, Top_journals.values)
plt.xlabel("Journal")
plt.ylabel("Number of Publications")
plt.title("Top Publishing Journals")
plt.show()
````

````python
#Generate a word cloud of paper titles
plt.figure(figsize=(10, 5))
plt.bar(freq_df['Word'], freq_df['Frequency'])
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.title("Word Frequency in Paper Titles")
plt.show()
````

````python
#Plot distribution of paper counts by source
plt.figure(figsize=(10, 5))
sns.countplot(x='source_x', data=df2)
plt.xlabel("Source")
plt.ylabel("Count")
plt.title("Distribution of Paper Counts by Source")
plt.show()
````

# **Part 4: Streamlit Application **

````python
# I’m setting up the title and description of my app
st.title("CORD-19 Data Explorer")
st.write("Simple exploration of COVID-19 research papers")

# I’m adding a slider so I can filter papers by publication year
year_range = st.slider("Select year range",df2['Publication_Year'].min(),df2['Publication_Year'].max(), (2020, 2021))

# I assume I already have a dataframe called df2 with a 'Publication_Year' column
# So I filter it based on the year range selected
filtered_df = df2[
    (df2['Publication_Year'] >= year_range[0]) &
    (df2['Publication_Year'] <= year_range[1])
]

# I want to visualize how many papers were published per year in the selected range
yearly_counts = filtered_df['Publication_Year'].value_counts().sort_index()

# I’m creating a bar chart using matplotlib
fig, ax = plt.subplots()
yearly_counts.plot(kind='bar', ax=ax)
ax.set_xlabel("Publication Year")
ax.set_ylabel("Number of Papers")
ax.set_title("Publications per Year")

# Finally, I display the chart in Streamlit
st.pyplot(fig)
````
